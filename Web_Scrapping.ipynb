```python
import requests
import bs4

url = input("Enter your url")
response = requests.get(url)
print(type(response))
```

```python
print(response.text)
```

```python
file_name = 'temp.html'
```

```python
bs = bs4.BeautifulSoup(response.text,'html.parser')
```

```python
format_text = bs.prettify()
```

```python
print(format_text)
```

```python
with open (file_name,"w+")as f:
    f.write(format_text)
```

```python
list_of_img = bs.find_all('img')
```

```python
print(list_of_img)
```
